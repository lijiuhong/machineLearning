## 一. scikit-learn的k-means聚类

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150,
                 n_features=2,
                 centers=3,
                 cluster_std=0.5,
                 shuffle=True,
                 random_state=0)

#创建散点图
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c="white", marker='o', edgecolor='black', s=50)
plt.grid()
plt.tight_layout()
plt.show()
```

### 算法实现：
        1.随机从样本中挑选k个重心作为初始聚类中心内；
        2.将每个样本分配到最近的重心μ(j),j∈{1,2，...,k}；
        3.把重心移到已分配样本的中心；
        4.重复步骤2和3，直到集群赋值不再改变，或者达到用户定义的容限或最大迭代数。

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
           init='random',
           n_init=10,
           max_iter=300,
           tol=1e-04,
           random_state=0)
y_km = km.fit_predict(X)
```

```python
#取出数据并画出对应的散点图
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='lightgreen', marker='s', edgecolor='black', label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='orange', marker='o', edgecolor='black', label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='lightblue', marker='v', edgecolor='black', label='cluster 3')

#画出中心点
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], s=250, marker='*', c='red', edgecolor='black', label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
print('Distortion:%.2f'%km.inertia_)
```

#### Elbow Method求解最佳聚类数

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
    
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

#### 基于k-means算法的聚类轮廓系数图

```python
km = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, random_state=0, tol=1e-04)
y_km = km.fit_predict(X)

import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i)/n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, edgecolor='none', color=color)
    yticks.append((y_ax_lower + y_ax_upper)/2.)
    y_ax_lower += len(c_silhouette_vals)

silhoouette_avg = np.mean(silhouette_vals)
plt.axvline(silhoouette_avg, color='red', linestyle='--')
plt.yticks(yticks, cluster_labels+1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.show()
```

## 层次聚类

```python
import pandas as pd
import numpy as np
np.random.seed(123)
variables = ['X', 'Y', 'Z']
labels = ['ID_0', 'ID_1', 'ID_2', 'ID_3', 'ID_4']
X = np.random.random_sample([5,3])*10
df = pd.DataFrame(X, columns=variables, index=labels)
df
```

```python
from scipy.spatial.distance import pdist, squareform
row_list = pd.DataFrame(squareform(pdist(df, metric='euclidean')), columns=labels, index=labels)
row_list
```

```python
y_distance = pdist(df, metric='euclidean')
y_distance
```

```python
squareform(y_distance)
```

```python
from scipy.cluster.hierarchy import linkage
row_clusters = linkage(pdist(df, metric='euclidean'), method='complete')
row_clusters2 = linkage(df.values, method='complete', metric='euclidean')
```

```python
pd.DataFrame(row_clusters, columns=['row label 1', 'row label 2', 'distance', 'no. of items in clust.'], 
             index=['cluster %d' %(i+1) for i in range(row_clusters.shape[0])])
```

```python
from scipy.cluster.hierarchy import dendrogram
row_dendr = dendrogram(row_clusters, labels=labels)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

### scikit-learn凝聚聚类方法

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
labels = ac.fit_predict(X)
print('cluster labels: %s' %labels)
```
