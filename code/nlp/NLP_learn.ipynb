### 2.2去除停用词

```python
import pandas as pd

stopwords_path = r"C:\Users\lijiuhong\Desktop\inews\stopwords-master\cn_stopwords.txt"
#with open(stopwords_path, 'r', encoding='UTF-8') as stop_obj :
    #conte = stop_obj.readlines()
temdf = pd.read_csv(stopwords_path, sep='_!_', names=['w'], encoding='utf-8', engine='python')
temdf.values
```

```python
len(temdf.values)
temdf.head(30)
```

```python
word_list = [word for word in jieba.lcut(temstr) if word not in temdf.values and len(word)>=2]

if '想' not in temdf.values :
    print('-----true')
else :
    print('false')
word_list
```

## 3.词云展示

### 3.1词频统计

```python
word_dic = {}
for word in word_list :
    if word in word_dic :
        word_dic[word] += 1
    else :
        word_dic[word] = 1

word_dic2 = {}
for word in word_dic.keys() :
    if word_dic[word] >=1 :
        word_dic2[word] = word_dic[word]
word_dic2
```

### 3.2使用pandas统计

```python
import pandas as pd
df = pd.DataFrame(word_list, columns=['word'])
df.head(30)
```

```python
result = df.groupby(['word']).size()
freqlist = result.sort_values(ascending=False)
freqlist[:20]
```

### 3.3使用NLTK统计词频

```python
import nltk
fdist = nltk.FreqDist(word_list)
fdist
```

## 4.关键词提取

### 4.1基于sklearn实现TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

txt_list = word_list
vectorize = CountVectorizer()
X = vectorize.fit_transform(txt_list)

transformer = TfidfTransformer()
tfidf = transformer.fit_transform(X)

tfidf.toarray()#转换为数组
tfidf.todense()
```
